```python
user = user[user.patron_profile_id.notna()]
user.shape
#remove the repeat rows with same email address and the patron_id is null
```

```python
def calculate_age(birth):
    today_d = datetime.datetime.now()
    age = today_d.year - birth + 1
    return age
```

```python
user['age'] = user['year_of_dob'].apply(lambda x: calculate_age(float(x)))
```

```python
user.title.unique()
```

```python
g1 = ['MDM','MS','MRS','MISS','DR & MRS']
g2 = ['MR','PRINCE','HE','M']
```

```python
def replacegender(astring):
    if astring in g1:
        return 'female'
    if astring in g2:
        return 'male'
    else:
        return 'NULL'
```

```python
user['gender'] = user['title'].apply(lambda x: replacegender(str(x)))
```

```python
df1 = user.nationality.value_counts().rename('cnt').reset_index()
df1['proportion'] = df1['cnt']*100/df1.cnt.sum()
df1.head(10)
```

```python
def replacenation(astring):
    if astring == 'SG':
        return 'SG'
    if astring == '-':
        return 'NULL'
    if astring == 'MY':
        return 'MY'
    if astring == 'CN':
        return 'CN'
    if astring == 'ID':
        return 'ID'
    if astring == 'IN':
        return 'IN'
    if astring == 'GB':
        return 'GB'
    if astring == 'PH':
        return 'PH' 
    else:
        return 'other'
```

```python
user['nationalitygroup'] = user['nationality'].apply(lambda x: replacenation(str(x)))
```

```python
df2 = user.nationalitygroup.value_counts().rename('cnt').reset_index()
df2['proportion'] = df2['cnt']*100/df2.cnt.sum()
df2
#the proportion below 2% will be grouped in 'other' label
```

```python
#nationality distribution 
fig = plt.figure(figsize=(12,4)) 
x = df2['index']
y = df2['cnt']
plt.bar(x,y)
for a,b in zip(x,y):
    plt.text(a, b+0.1, '%.0f' % b, ha='center', va= 'bottom',fontsize=10)
plt.show()
```

```python
df3 = user.country_of_residense.value_counts().rename('cnt').reset_index()
df3['proportion'] = df3['cnt']*100/df3.cnt.sum()
df3.head(10)
```

```python
def replaceresidence(astring):
    if astring == 'SG':
        return 'SG'
    if astring == '-':
        return 'NULL'
    if astring == 'MY':
        return 'MY'
    if astring == 'ID':
        return 'ID'
    else:
        return 'other'
```

```python
user['residensegroup'] = user['country_of_residense'].apply(lambda x: replaceresidence(str(x)))
```

```python
df4 = user.residensegroup.value_counts().rename('cnt').reset_index()
df4['proportion'] = df4['cnt']*100/df4.cnt.sum()
df4
#the proportion below 1% will be grouped in 'other' label
```

```python
#residence distribution 
fig = plt.figure(figsize=(8,4)) 
x = df4['index']
y = df4['cnt']
plt.bar(x,y)
for a,b in zip(x,y):
    plt.text(a, b+0.1, '%.0f' % b, ha='center', va= 'bottom',fontsize=10)
plt.show()
```

```python
def replaceage(afloat):
    if afloat <18:
        return '<18'
    if afloat >=18 and afloat <25:
        return '18-25'
    if afloat >=25 and afloat <30:
        return '25-30'
    if afloat >=30 and afloat <40:
        return '30-40'
    if afloat >=40 and afloat <50:
        return '40-50'
    if afloat >=50 and afloat <60:
        return '50-60'
    if afloat >=60:
        return '>=60'
    else:
        return 'NULL'
```

```python
user['agegroup'] = user['age'].apply(lambda x: replaceage(float(x)))
user['agegroup']
```

```python
x1 = user.groupby('agegroup').patron_profile_id.nunique().rename('cnt').sort_values(ascending = False).reset_index()
x1
```

```python
#age distribution
fig = plt.figure(figsize=(12,4)) 
x = x1['agegroup']
y = x1['cnt']
plt.bar(x,y)
for a,b in zip(x,y):
    plt.text(a, b+0.1, '%.0f' % b, ha='center', va= 'bottom',fontsize=10)
plt.show()
```

```python
x2 = user.groupby('gender').patron_profile_id.nunique().rename('cnt').sort_values(ascending = False).reset_index()
x2
```

```python
#gender distribution 
fig = plt.figure(figsize=(4,4)) 
x = x2['gender']
y = x2['cnt']
plt.bar(x,y)
for a,b in zip(x,y):
    plt.text(a, b+0.1, '%.0f' % b, ha='center', va= 'bottom',fontsize=10)
plt.show()
```
